# Creating dataframes for Genres, Tags, Production companies and relation tables

```python
#Creating the table for Genres
import json
genres_df=pd.DataFrame(movies_df,columns=['genres'])
genres = pd.DataFrame(columns=['genre_id', 'genre_name'])
i=1
for gr in genres_df.itertuples():
    gen=gr[1]
    d=json.loads(gen)
    for dg in d:
        name= dg['name']
        id_g=dg['id']
        #print name,id_g
        genres.loc[id_g]=[id_g,name]
print genres
```

```python
#Creating the table for Tags
keywords_df=pd.DataFrame(movies_df,columns=['keywords'])
Tags = pd.DataFrame(columns=['Tag_id', 'Tag_name'])
i=1
for key in keywords_df.itertuples():
    k=key[1]
    d=json.loads(k)
    for dk in d:
        name= dk['name']
        id_k=dk['id']
        Tags.loc[id_k]=[id_k,name]
print Tags
```

```python
#Creating the relation table for Movies and Genres
import json
genres_df=pd.DataFrame(movies_df,columns=['id','genres'])
genre_movies = pd.DataFrame(columns=['movie_id','genre_id'])
i=1
for gr in genres_df.itertuples():
    gen=gr[2]
    id_movie=gr[1]
    d=json.loads(gen)
    for dg in d:
        id_g=dg['id']
        genre_movies.loc[str(i)]=[id_movie,id_g]
        i=i+1
print genre_movies
```

```python
#Creating the relation table for Movies and Tags
keywords_df=pd.DataFrame(movies_df,columns=['keywords','id'])
tags_movies = pd.DataFrame(columns=['movie_id','tag_id'])
i=1
for key in keywords_df.itertuples():
    k=key[1]
    id_movie=key[2]
    d=json.loads(k)
    for dk in d:
        id_k=dk['id']
        tags_movies.loc[str(i)]=[id_movie,id_k]
        i=i+1
print tags_movies
```

```python
#Creating the table for Production Companies
pc_df=pd.DataFrame(movies_df,columns=['production_companies'])
productionCompanies = pd.DataFrame(columns=['prodComp_id', 'company_name'])
i=1
for p in pc_df.itertuples():
    pc=p[1]
    prod=json.loads(pc)
    for pdc in prod:
        name= pdc['name']
        id_p=pdc['id']
        productionCompanies.loc[id_p]=[id_p,name]
print productionCompanies
```

```python
#Creating the relation table for Movies and Production companies
production_Movies_df=pd.DataFrame(movies_df,columns=['production_companies','id'])
movies_prodCompanies = pd.DataFrame(columns=['movie_id','prodComp_id'])
i=1
for pcm in production_Movies_df.itertuples():
    prod_mov=pcm[1]
    id_movie=pcm[2]
    pm=json.loads(prod_mov)
    for p_m in pm:
        id_pm=p_m['id']
        movies_prodCompanies.loc[str(i)]=[id_movie,id_pm]
        i=i+1
print movies_prodCompanies
```

# Making the connection to the sqlite database

```python
import sqlite3
conn = sqlite3.connect('Shruthi_TMDB.db') 
c = conn.cursor()
conn.text_factory = lambda x: unicode(x, 'utf-8', 'ignore')
```

# Creating the tables in the database and inserting values from the dataframes into the respective tables

```python
c.execute("Create table Movies(id int PRIMARY KEY,title varchar(50),original_language varchar(20),release_date datetime,popularity float,budget longint,revenue longint,runtime float)")
movies.to_sql('Movies',con=conn,if_exists='append',index=False)
```

```python
c.execute("Create table Genres(genre_id int PRIMARY KEY,genre_name varchar(20))")
genres.to_sql('Genres',con=conn,if_exists='append',index=False)
```

```python
c.execute("Create table Tags(Tag_id int PRIMARY KEY,Tag_name varchar(50))")
Tags.to_sql('Tags',con=conn,if_exists='append',index=False)
```

```python
c.execute("Create table Movie_Genres(movie_id int,genre_id int,PRIMARY KEY(movie_id,genre_id))")
genre_movies.to_sql('Movie_Genres',con=conn,if_exists='append',index=False)
```

```python
c.execute("Create table Movie_Tags(movie_id int,tag_id int,PRIMARY KEY(movie_id,tag_id))")
tags_movies.to_sql('Movie_Tags',con=conn,if_exists='append',index=False)
```

```python
c.execute("Create table ProductionCompanies(prodComp_id int PRIMARY KEY,company_name varchar(50))")
productionCompanies.to_sql('ProductionCompanies',con=conn,if_exists='append',index=False)
```

```python
c.execute("Create table Movies_ProdCompanies(movie_id int,prodComp_id int,PRIMARY KEY(movie_id,prodComp_id))")
movies_prodCompanies.to_sql('Movies_ProdCompanies',con=conn,if_exists='append',index=False)
```

```python
c.execute("Create table Actors(actor_id int PRIMARY KEY, name varchar(50))")
actor_df.to_sql('Actors',con=conn,if_exists='append',index=False)
```

```python
c.execute("Create table Cast(movie_id int, actor_id int, character varchar(50),PRIMARY KEY(movie_id,actor_id,character))")
df_cast.to_sql('Cast',con=conn,if_exists='append',index=False)
```

# All tables are Normalized

    The primary key of the Movies table is ID. Every movie has a unique ID, used to distinguish it. The table has atomic values in every column, and no two columns store the same values. All the non-key attributes are completely dependent on the movie ID. There are no transitive dependencies.

    The primary key of the Genres table is genre_id. Every genre has a unique ID, used to distinguish it. The table has atomic values in every column, and no two columns store the same values. The genre_name is completely dependent on the genre ID. 

    The primary key of the Tags table is Tag_id. Every tag/keyword has a unique ID, used to distinguish it. The table has atomic values in every column, and no two columns store the same values. The Tag name is completely dependent on the movie ID.

    The primary key of the Movie_genres table is a composite key of movie_id and genre_id. The pair of movie and a particular genre is unique. The table has atomic values in every column, and no two columns store the same values.

    The primary key of the Movie_Tags table is a composite key of movie_id and tag_id. The pair of movie and a particular Tag/keyword is unique. The table has atomic values in every column, and no two columns store the same values.

    The primary key of the ProductionCompanies table is the ID of the production company prodComp_id. Every company has a unique ID, used to distinguish it. The table has atomic values in every column, and no two columns store the same values. The company name is completely dependent on the production company ID.

    The primary key of the Movies_ProdCompanies table is a composite key of movie ID and production company ID. Every movie has a unique ID, used to distinguish it. The table has atomic values in every column, and no two columns store the same values.All the non-key attributes are completely dependent on the movie ID. There are no transitive dependencies.

    The primary key of the Movies table is ID. The pair of movie and a particular Tag/keyword is unique . The table has atomic values in every column, and no two columns store the same values.

    The primary key of the Actors table is the actor ID. Every actor has a unique ID, used to distinguish him/her. The table has atomic values in every column, and no two columns store the same values. 

    The primary key of the Cast table is a composite key of movie ID, actor ID and character name. The combination movie id, actor id and the character name of the actor in that movie are unique. If an actor plays a double role in the same movie, he would only have a different character name.  The table has atomic values in every column, and no two columns store the same values.
